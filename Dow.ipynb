```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
import markdown
import statistics
from statistics import NormalDist
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import glm
```

```python
dow = pd.read_csv("Dow2.dat", sep="\s+")
dow_model = smf.ols(formula="nextopen ~ open + high + low + close", data=dow).fit()
print(dow_model.summary(title="NextOpen - predicted by Open, High, Low, Close data"))
print(dow_model.params)
```

```python
plt.figure(figsize=(12,12), dpi=100)
for i in range(len(dow)):
    plt.scatter(dow['open'][i],dow['nextopen'][i], marker='o', color='#0077dd', alpha=1.0, s=30)
    plt.annotate(dow['stock'][i], (dow['open'][i],dow['nextopen'][i]), xytext=(6, -5), textcoords='offset points')
plt.xlabel("Weekly Opening Price Moving Average, as of previous week")
plt.ylabel("Opening price, end of dataset")
plt.title("Relationship of weekly opening price moving average to Opening Price at end of dataset")
plt.rc('grid', linestyle="-", color='#bbbbbb')
plt.grid()
plt.show()
```
